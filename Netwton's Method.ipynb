# Newtons Method

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def fx(xk):
    return (7*xk**4)+(3*xk**3)+(2*xk**2)+(9*xk)+4

def dfx(xk):
    return (28*xk**3)+(9*xk**2)+(4*xk)+9
```

```python
xk = 0
count = 0
while(True):

    xk = xk - fx(xk)/dfx(xk)
    print(xk)
    count += 1
    if fx(xk) == 0:
        break
print("Iterations: ", count)
```

```python
def fx(xk):
    return (xk-2)**4 + (xk-2)**5

def dfx(xk):
    return 4*(xk-2)**3 + 5*(xk-2)**4
```

```python
xk = 3
count = 0
k_label = []
x1_label = []
error_label, fxk_label = [], []

while(True):
    
    prev = xk
    xk = xk - fx(xk)/dfx(xk)
        
    #print(xk)
    k_label.append(count)
    x1_label.append(xk)
    fxk_label.append(fx(xk))
    error_label.append(abs(prev-xk))
    
    count += 1
    if np.round(fx(xk), 10) == 0:
        break
    
print("Iterations: ", count)
df = [k_label, x1_label, fxk_label, error_label]
df = pd.DataFrame(df, index=['k', 'xk', 'f(xk)', 'x-x*']).T
df
```

```python
plt.plot(fxk_label)
plt.title("Convergence Graph")
plt.ylabel("f(xk)")
plt.xlabel("k")
plt.show()
```

# Multivariate

```python
def f1(x1, x2):
    return (3*x1*x2) + (7*x1) + (2*x2) - 3

def f2(x1, x2):
    return (5*x1*x2) - (9*x1) - (4*x2) + 6
```

```python
def df1x1(x1, x2):
    return (3*x2) + 7

def df1x2(x1, x2):
    return (3*x1) + 2

def df2x1(x1, x2):
    return (5*x2) - 9

def df2x2(x1, x2):
    return (5*x1) - 4
```

```python
def f(x1, x2):
    return np.array([(3*x1*x2) + (7*x1) + (2*x2) - 3, (5*x1*x2) - (9*x1) - (4*x2) + 6])
```

```python
x0 = np.array([1, 2])
k = 0

k_label = []
x1_label, x2_label = [], []
norm_label, error_label = [], []

x1, x2 = x0[0], x0[1]

while(True):
    
    mat1 = np.array([[df1x1(x1, x2), df2x1(x1, x2)], [df1x2(x1, x2), df2x2(x1, x2)]])
    mat1 = np.linalg.inv(mat1)
    mat2 = np.array([f1(x1, x2), f2(x1, x2)])

    if k == 0:
        x1, x2 = x0 - np.dot(mat2, mat1.T)
    else:
        prev = [x1, x2]
        x1, x2 = np.array([x1, x2]) - np.dot(mat2, mat1)
        
    norm = np.linalg.norm(f(x1, x2))
    error = np.linalg.norm(np.array([x1, x2])-np.array(prev))
    
    k_label.append(k)
    x1_label.append(x1)
    x2_label.append(x2)
    norm_label.append(norm)
    error_label.append(error)

    k += 1
    if np.round(norm, 20) == 0 or count == 1000:
        break
    
print("Iterations: ", k)
df = [k_label, x1_label, x2_label, norm_label, error_label]
df = pd.DataFrame(df, index=['k', 'x1', 'x2', '||f||2', '||x-x*||2']).T
df
```
